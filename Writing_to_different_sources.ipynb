```python
import pandas as pd
```

```python
df = pd.read_csv('survey_results_public.csv')
df.head(10)
```

```python
filt=df['Country'] == 'India'
indian_df=df[filt]
indian_df.head(10)
```

```python
# saves the filtered dataframe to a new file with a tab separator
indian_df.to_csv('indian_survey.tsv', sep='\t')
```

```python
indian_df.to_excel('indian_survey.xlsx', index=False)
```

```python
indian_df.to_json('indian_survey.json', orient='records', lines=True)
```
