```python
employment_respondents = df['Employment'].value_counts()
employment_respondents
```

```python
employers_uses_gemini=employment_group['AIAgentExternal'].apply(lambda x: x.str.contains('Gemini').sum())
employers_uses_gemini
```

```python
new_df=pd.concat([employment_respondents, employers_uses_gemini],axis=1,sort=False)
new_df.rename(columns={'Employment':'Respondents', 'AIAgentExternal':'Employers_Uses_Gemini','count':'Number of Respondents'}, inplace=True)
new_df
```

```python
new_df['Percentage_Employers_Uses_Gemini'] = (new_df['Employers_Uses_Gemini'] / new_df['Number of Respondents']) * 100
new_df
```

```python
new_df.sort_values(by='Percentage_Employers_Uses_Gemini', ascending=False, inplace=True)
new_df
```
